```python
from langchain_huggingface import HuggingFaceEndpoint, ChatHuggingFace
from langchain_core.tools import tool
import requests
```

```python
from langchain_community.tools import DuckDuckGoSearchRun

search_tool = DuckDuckGoSearchRun()

results = search_tool.invoke('top news in India')
```

```python
results
```

```python
@tool
def get_weather_data(city: str) -> str:
    """
    this function fetches the current wether data for a given city
    """
    url = f"http://api.weatherstack.com/current?access_key=xxxxxxxxxxxxxxx&query={city}"
    response = requests.get(url)
    return response.json()
```

```python
llm = HuggingFaceEndpoint(
    repo_id = "deepseek-ai/DeepSeek-V3.2"
)
model = ChatHuggingFace(llm=llm)
```

```python
model.invoke('hii')
```

```python

from langchain_classic.agents import AgentExecutor, create_react_agent
from langchain_classic import hub
```

```python
prompt = hub.pull("hwchase17/react")
```

```python
agent = create_react_agent(
    llm = model,
    tools = [search_tool, get_weather_data],
    prompt = prompt
)
```

```python
agent_executor = AgentExecutor(
    agent = agent,
    tools = [search_tool, get_weather_data],
    verbose = True
)
```

```python
response = agent_executor.invoke({"input": "what is the capital of meghalaya and find its current weather condition"})
print(response)
```
