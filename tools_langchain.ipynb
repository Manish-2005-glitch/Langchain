Custom_tools

@tool decorator

```python
from langchain_core.tools import tool
@tool
def multiply(a:int, b:int) -> int:
    """Multiply two numbers"""
    return a * b
```

```python

multiply.invoke({"a":4,"b":5})
```

```python
print(multiply.name)
print(multiply.description)
print(multiply.args)
```

```python
from langchain_core.tools import StructuredTool
from pydantic import BaseModel, Field
```

```python
class MultiplyInput(BaseModel):
    a: int = Field(required = True, description = "The first number")
    b: int = Field(required = True, description= " The second number")
```

```python
def multiply_func(a:int, b:int) -> int:
    return a*b
#here no need of type hint actually 
```

```python
multiply_tool = StructuredTool.from_function(
    func = multiply_func,
    name = "multiply",
    description = "Multiply two numbers",
    args_schema = MultiplyInput
)
```

```python
result = multiply_tool.invoke({"a":5,"b":5})
```

```python
print(result)
print(multiply_tool.name)
print(multiply_tool.description)
print(multiply_tool.args)
```

```python
from langchain_core.tools import BaseTool
from typing import Type
```

```python
class MultiplyInput(BaseModel):
    a:int = Field(required =True, description = "First Number")
    b: int = Field(required = True, description = "Second Number")
```

```python
class MultiplyTool(BaseTool):
    name: str = "multiply"
    description: str = "multiply two numbers"
    
    args_schema: Type[BaseModel] = MultiplyInput
    
    def _run(self, a:int, b:int) -> int:
        return a*b
```

```python
multiply_tool = MultiplyTool()
```

```python
result = multiply_tool.invoke({'a':3, 'b':4})

print(result)

print(multiply_tool.name)
print(multiply_tool.description)
print(multiply_tool.args)
```

```python
@tool
def add(a:int, b:int)-> int:
    """Add two numbers"""
    return a + b

@tool
def multiply(a:int, b:int)->int:
    """Multiply two numbers"""
    return a*b
    
```

```python
class MathToolkit:
    def get_tools(self):
        return[add,multiply]
```

```python
toolkit=MathToolkit()
tools = toolkit.get_tools()

for tool in tools:
    print(tool.name, "=>", tool.description)
```
